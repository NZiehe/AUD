```c++
#include <algoviz/SVG.hpp>
#include <string>
#include <iostream>
#include <cmath>
#include <math.h>
using namespace std;
```

```c++
string key = "";
int wait = 200;
int dx = -10;
int dy = 0;
int size = 5;

SVG canvas = SVG(400,400);
Rect border = Rect(0, 0, 400, 400, &canvas);
border.setFill("white");
Circle snake[50];
Text points = Text("Points: " + to_string(size -5),10,20,&canvas);

for(int i = 0; i < size; i++){
    
    snake[i].addTo(&canvas);
    snake[i].setRadius(10);
    snake[i].setFill("green");
    snake[i].moveTo(200 + i*10, 200);
}
snake[0].setFill("red");

Circle snacky = Circle((rand() % 181 + 10), (rand() % 181 + 10), 10, &canvas);
snacky.setFill("yellow");
```

```c++
bool collisionCircle(int c1x, int c1y, int c2x, int c2y, int r1, int r2){
    
    double x = c2x - c1x;
    double y = c2y - c1y;
    
    if(sqrt(pow(x,2) + pow(y,2)) < r1 + r2){
        return true;
    }
    return false;
}
```

```c++
void keyCheck(){
    if ( key == "ArrowUp") {
        dx = 0;
        dy = -10;
    } else if ( key == "ArrowDown") {
        dx = 0;
        dy = 10;
    } else if ( key == "ArrowLeft") {
        dx = -10;
        dy = 0;
    } else if ( key == "ArrowRight") {
        dx = 10;
        dy = 0;
    }
}
```

```c++
void snack(){
    if(collisionCircle(snake[0].getX(), snake[0].getY(), snacky.getX(), snacky.getY(), snake[0].getRadius(), snacky.getRadius())){
        snacky.moveTo((rand() % 181 + 10), (rand() % 181 + 10));
        size++;
        if(size <= 50){
            snake[size-1].addTo(&canvas);
            snake[size-1].setRadius(10);
            snake[size-1].setFill("green");
 
            if ( key == "ArrowUp") {
                snake[size-1].moveTo(snake[size-2].getX(), snake[size-2].getY()-10);
                
            } else if ( key == "ArrowDown") {
                snake[size-1].moveTo(snake[size-2].getX(), snake[size-2].getY()+10);
                
            } else if ( key == "ArrowLeft") {
                snake[size-1].moveTo(snake[size-2].getX()-10, snake[size-2].getY());
                
            } else if ( key == "ArrowRight") {
                snake[size-1].moveTo(snake[size-2].getX()+10, snake[size-2].getY());
            }
        }
        points.setText("Points: " + to_string(size - 5));
    }
}
```

```c++
void wallCollision(){
    
    if(snake[0].getX() + 10 >= 400){
        canvas.drawText("GAME OVER",100,200,20,"sans-serif");
        key = "q";
    }
    if(snake[0].getX() - 10 <= 0){
        canvas.drawText("GAME OVER",100,200,20,"sans-serif");
        key = "q";
    }
    if(snake[0].getY() + 10 >= 400){
        canvas.drawText("GAME OVER",130,200,20,"sans-serif");
        key = "q";
    }
    if(snake[0].getY() - 10 <= 0){
        canvas.drawText("GAME OVER",100,200,20,"sans-serif");
        key = "q";
    }
    return false;
}
```

```c++
int counter = 0;
Text t = Text("" + to_string(counter),300,300,&canvas);
while(key != "q"){
    
    key = canvas.lastKey();
    keyCheck();
    
    //movement
    for(int i = size - 1; i > 0; i--){
        snake[i].moveTo(snake[i-1].getX(), snake[i-1].getY());
    }
    snake[0].moveTo(snake[0].getX() + dx, snake[0].getY() + dy);
    counter++;
    t.setText("" + to_string(counter));
    snack();
    wallCollision();
    
    for(int i = size-1; i > 0; i--){
        if(collisionCircle(snake[0].getX(),snake[i].getX(),snake[0].getY(),snake[i].getY(),5,5)){
            key ="q";
            canvas.drawText("WTF",100,200,20,"sans-serif");
        }
    }
                                       
    AlgoViz::sleep(wait);
}
```
