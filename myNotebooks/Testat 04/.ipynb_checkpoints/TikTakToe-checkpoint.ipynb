```c++
#include <iostream>
#include <algoviz/SVG.hpp>
#include <string>
using namespace std;
```

```c++
SVG canvas = SVG(300,400);
Rect field[3][3];
char status[3][3];
string key = "l";
char player = '1';
Text turn = Text("",20,380,&canvas);
MouseState mouse;
Text message = Text("",70,360,&canvas);

for(int i = 0; i < 3; i++){
    for(int j = 0; j < 3; j++){
        status[i][j] = '0';
    }
}

for(int i = 0; i < 3; i++){
    for(int j = 0; j < 3; j++){
        field[i][j].addTo(&canvas);
        field[i][j].setWidth(100);
        field[i][j].setHeight(100);
        field[i][j].moveTo((i*100),(j*100));
        field[i][j].setFill("white");
    }
}
```

```c++
bool winCheck(){
    
    if((status[0][0] == '1') && (status[1][0] == '1') && (status[2][0] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[0][0] == '2') && (status[1][0] == '2') && (status[2][0] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[0][1] == '1') && (status[1][1] == '1') && (status[2][1] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[0][1] == '2') && (status[1][1] == '2') && (status[2][1] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[0][2] == '1') && (status[1][2] == '1') && (status[2][2] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[0][2] == '2') && (status[1][2] == '2') && (status[2][2] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[0][0] == '1') && (status[0][1] == '1') && (status[0][2] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[0][0] == '2') && (status[0][1] == '2') && (status[0][2] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[1][0] == '1') && (status[1][1] == '1') && (status[1][2] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[1][0] == '2') && (status[1][1] == '2') && (status[1][2] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[2][0] == '1') && (status[2][1] == '1') && (status[2][2] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[2][0] == '2') && (status[2][1] == '2') && (status[2][2] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[0][0] == '1') && (status[1][1] == '1') && (status[2][2] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[0][0] == '2') && (status[1][1] == '2') && (status[2][2] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else if((status[0][2] == '1') && (status[1][1] == '1') && (status[2][0] == '1')){
        message.setText("!!!Spieler 1 hat gewonnen!!!");
        return true;
        
    }else if((status[0][2] == '2') && (status[1][1] == '2') && (status[2][0] == '2')){
        message.setText("!!!Spieler 2 hat gewonnen!!!");
        return true;
        
    }else{
        bool draw = true;
        
        for(int i = 0; i < 3; i++){
            for(int j = 0; j < 3; j++){
                if(status[i][j] == '0'){
                    draw = false;
                }
            }
        }
        if(draw){
            message.setText("!!!DRAW!!!");
        }
        return draw;
    }
    return false;
}
```

```c++
bool action(){
    int x = mouse.x();
    int y = mouse.y();
    int indexX = -1;
    int indexY = -1;
    
    switch((x / 100)){
        case 0: indexX = 0;
            break;
        case 1: indexX = 1;
            break;
        case 2: indexX = 2;
            break;
    }
    
    switch((y / 100)){
        case 0: indexY = 0;
            break;
        case 1: indexY = 1;
            break;
        case 2: indexY = 2;
            break;
    }
    
    if(status[indexX][indexY] == '0'){
        if(player == '1'){
            field[indexX][indexY].setFill(0,0,255,1);
            status[indexX][indexY] = '1';
            player++;
            return true;
        }else{
            field[indexX][indexY].setFill(255,0,0,1);
            status[indexX][indexY] = '2';
            player--;
            return true;
        }
    }else{
        
        return false;
    }
    
    return false;
}
```

```c++
while(key != "q"){
    
    key = canvas.lastKey();
    
    do{
        if(player == '1'){
            turn.setText("Spieler " + to_string(player - 48) + " ist am Zug (Farbe: Blau)");
        }else{
            turn.setText("Spieler " + to_string(player - 48) + " ist am Zug (Farbe Rot)");
        }
        
        canvas.waitForClick();
        mouse = canvas.getMouseState();
        
    }while(action() == false);
    
    if(winCheck()){
        key = "q";
        turn.setText("");
    }
    if(key == "q"){
        turn.setText("Spiel wurde durch 'q' abgebrochen!!");
    }
}
```
