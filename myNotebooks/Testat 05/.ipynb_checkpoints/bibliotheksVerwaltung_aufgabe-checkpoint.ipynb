## Anforderungen

### Die Klasse Bibliothek
- verwaltet Bücher und Spiele in einer Liste,
- enthält mindestens folgende Operationen:
    - *void addBuch(string titel, string autor, int eJahr)* (erstellt ein Buch und fügt dieses hinzu),
    - *void addSpiel(string titel, int szahl, int alter)* (erstellt ein Spiel und fügt dieses hinzu),
    - *bool entferneItem(int id)* (Löscht das Item mit der id),
    - Suchen und Ausgabe mit bestimmten Kriterien:
        - *void sucheTitel(string titel)*,
        - *void sucheAutor(string autor)*,
        - *void sucheSpielerzahl(int sZahl)*,
        - *void sucheID(int id)*,
    - *bool verleiheItem(int id)* (Verleiht das Item mit der id),
    - *bool gebeItemZurueck(int id)* (Gibt das Item mit der id zurück),
    - *bool pruefeVerleihstatus(int id)* (prüft, ob das Item mit der id verliehen ist),
    - *void druckeInventar()* (gibt alle in der Bibliothek gespeicherten Items aus).
    
### Die Klasse Item
- hat Variablen für folgende Eigenschaften:
    - Typ des Items (Buch oder Spiel),
    - einmalige ID,
    - Titel des Items,
    - Autor (wird nur gesetzt, falls das Item ein Buch ist),
    - Erscheinungsjahr (wird nur gesetzt, falls das Item ein Buch ist),
    - Spieleranzahl (wird nur gesetzt, falls das Item ein Spiel ist),
    - empfohlenes Alter (wird nur gesetzt, falls das Item ein Spiel ist),
    - Ausleihstatus,
- hat mindestens einen Konstruktor für ein neues Buch und ein neues Spiel,
- hat eine *toString-Operation*, die erkennt, ob es sich um ein Buch oder ein Spiel handelt und nur die relevanten Informationen zurückgibt,
- erfüllt das Prinzip der Kapselung.

### Keine Veränderung des vorgegebenen Codes!

*Hinweis*: Bitte keine Leerzeichen in Strings eingeben und denken Sie an geeignete Fehlerbehandlung! Bei der Fehlerbehandlung ist allerdings die Sicherstellung des Typs nicht notwendig. Gehen Sie hierbei ausnahmsweise davon aus, dass der Nutzer bspw. eine Zahl eingibt, wenn diese gefordert wird.

## Vorgegebener Code

```c++
#include <iostream>
#include <string>
#include <list>
#include "lsg/Bibliothek.hpp"
using namespace std;

Bibliothek bib = Bibliothek();
int eingabeZahl;
string titel;
string autor;
int eJahr;
int sZahl;
int alter;
int id;
bool verliehen;
bool itemGeloescht;

do {
    cout << endl;
    cout << "*******************************************************************************" << endl;
    cout << endl;
    cout << "Bitte wählen Sie eine Aktion:" << endl;
    cout << "1: Item hinzufügen," << endl;
    cout << "2: Item löschen," << endl;
    cout << "3: Item suchen," << endl;
    cout << "4: Ausleihstatus prüfen," << endl;
    cout << "5: Item ausleihen," << endl;
    cout << "6: Item zurücknehmen," << endl;
    cout << "7: Inventar der Bibliothek ansehen," << endl;
    cout << "10: beenden (Achtung: löscht alle Eingaben!)." << endl;
    cout << endl;
    
    cin >> eingabeZahl;
    
    cout << endl;
    
    switch(eingabeZahl) {
        case 1: {
            char eingabeZeichen;
            do {
                cout << "Möchten Sie ein Buch(b) oder ein Spiel(s) hinzufügen? " << endl;
                cin >> eingabeZeichen;
            } while(eingabeZeichen != 's' && eingabeZeichen != 'b');
            
            cout << endl; 
            
            switch(eingabeZeichen) {
                case 's': 
                    
                    cout << "Bitte Spezifikationen des Spiels eingeben: " << endl;
                    
                    do {
                        cout << "Titel: " << endl;
                        getline(cin, titel);
                    } while(titel.length() < 1);
                    
                    cout << endl;

                    do{
                        cout << "Spieleranzahl: " << endl;
                        cin >> sZahl;
                    } while(sZahl < 1);
                    
                    cout << endl;

                    do{
                        cout << "Empfohlenes Alter: " << endl;
                        cin >> alter;
                    } while(alter < 0 || alter > 100);
                    
                    bib.addSpiel(titel, sZahl, alter);
                    
                    cout << endl;
                    cout << "Das Spiel " << titel << " wurde hinzugefügt.";
                    cout << endl;
                    
                    break;
                
                case 'b': 
                    
                    cout << "Bitte Spezifikationen des Buches eingeben: " << endl;
                    
                    cout << endl;
                    do {
                        cout << "Titel: " << endl;
                        getline(cin, titel);
                    } while(titel.length() < 1);
                    
                    cout << endl;

                    do {
                        cout << "Autor: " << endl;
                        getline(cin, autor);
                    } while(autor.length() < 1);
                    
                    cout << endl;

                    do{
                        cout << "Erscheinungsjahr: " << endl;
                        cin >> eJahr;
                    } while(eJahr > 2021);
                    
                    bib.addBuch(titel, autor, eJahr);
                    
                    cout << endl;
                    cout << "Das Buch " << titel << " wurde hinzugefügt.";
                    cout << endl;
                    
                    break;
                
                default:
                    cout << endl;
                    cout << "Fehler, hinzufügen gescheitert!" << endl;
                    cout << endl;
                    
                    break;
            }; break;
        }
            
        case 2: {
            
            cout << endl;
            do{
                cout << "Bitte die ID des zu löschenden Items eingeben: " << endl;
                cin >> id;
            } while(id < 0);
            
            itemGeloescht = bib.entferneItem(id);
            
            cout << endl;
            if(itemGeloescht) {
                cout << "Item gelöscht!" << endl;
            } else {
                cout << "Fehler, löschen gescheitert!" << endl;
            }
            cout << endl;
            
            break; 
        }
        
        case 3: {
            
            char eingabeZeichen;
            do {
                cout << "Möchten Sie nach einem Titel(t), einem Autor(a), einer Spieleranzahl(s) oder einer ID(i) suchen? " << endl;
                cin >> eingabeZeichen;
            } while(eingabeZeichen != 't' && eingabeZeichen != 'a' && eingabeZeichen != 's' && eingabeZeichen != 'i');
            
            cout << endl; 
            
            switch(eingabeZeichen) {
                case 't': 
                                        
                    do {
                        cout << "Bitte den Titel des Items eingeben: " << endl;
                        getline(cin, titel);
                    } while(titel.length() < 1);
                    
                    bib.sucheTitel(titel);
                    
                    cout << endl;
                    
                    break;
                
                case 'a': 
                                        
                    do {
                        cout << "Bitte den Autoren des Items eingeben: " << endl;
                        getline(cin, autor);
                    } while(autor.length() < 1);
                    
                    bib.sucheAutor(autor);
                    
                    cout << endl;
                    
                    break;
                    
                case 's': 
                                        
                    do {
                        cout << "Bitte die Spieleranzahl des Items eingeben: " << endl;
                        cin >> sZahl;
                    } while(sZahl < 0);
                    
                    bib.sucheSpielerzahl(sZahl);
                    
                    cout << endl;
                    
                    break;
                    
                case 'i': 
                                        
                    do {
                        cout << "Bitte die ID des Items eingeben: " << endl;
                        cin >> id;
                    } while(id < 0);
                    
                    bib.sucheID(id);
                    
                    cout << endl;
                    
                    break;
                
                default:
                    cout << endl;
                    cout << "Fehler, hinzufügen gescheitert!" << endl;
                    cout << endl;
                    
                    break;
            }; break;
        }
        
        case 4: {
            
            cout << endl;
            do {
                cout << "Bitte die ID des zu prüfenden Items eingeben: " << endl;
                cin >> id;
            } while(id < 0);
            
            verliehen = bib.pruefeVerleihstatus(id);
            
            cout << endl;
            if(verliehen) {
                cout << "Das Item ist verliehen!" << endl;
            } else {
                cout << "Das Item ist nicht verliehen!" << endl;
            }
            cout << endl;
            
            break;
        }
            
        case 5: {
            
            cout << endl;
            do {
                cout << "Bitte die ID des auzuleihenden Items eingeben: " << endl;
                cin >> id;
            } while(id < 0);
            
            verliehen = bib.verleiheItem(id);
            
            if(verliehen) {
                cout << endl;
                cout << "Das Item wurde ausgeliehen!" << endl;
                cout << endl;
            } else {
                cout << endl;
                cout << "Fehler, Item nicht vorhanden oder verliehen!" << endl;
                cout << endl;
            }
            
            break; 
        }
            
        case 6: {
            
            cout << endl;
            do {
                cout << "Bitte die ID des zurückzugebenden Items eingeben: " << endl;
                cin >> id;
            } while(id < 0);
            
            verliehen = bib.gebeItemZurueck(id);
            
            if(verliehen) {
                cout << endl;
                cout << "Das Item wurde zurückgenommen!" << endl;
                cout << endl;
            } else {
                cout << endl;
                cout << "Fehler, Item nicht vorhanden oder nicht verliehen!" << endl;
                cout << endl;
            }
            
            break; 
        }
            
        case 7: {
            
            cout << "Es sind aktuell folgende Bücher und Spiele verfügbar: " << endl << endl;
            
            bib.druckeInventar();
            
            cout << endl;
            
            break;
        }
            
        case 10: {
            break;
        }
            
        default: {
            cout << "Fehler, bitte korrekte Eingabe!" << endl;
            break;
        }
    }
  
} while (eingabeZahl != 10);

cout << "Bibliotheksverwaltung beendet." << endl;
```
